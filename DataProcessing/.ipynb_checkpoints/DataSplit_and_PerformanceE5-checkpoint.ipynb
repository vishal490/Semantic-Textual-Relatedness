### Perform data split

```python
def random_sample(df, val_split=0.1):
    # Creating a dataframe with 50%
    # values of original dataframe
    val_df = df.sample(frac = val_split)
 
    # Creating dataframe with 
    # rest of the 50% values
    train_df = df.drop(val_df.index)

    return train_df, val_df
```

```python
train_df = pd.DataFrame()
val_df = pd.DataFrame()

for lang, lang_df in df.groupby(by="Lang"):
    train_lang, val_lang = random_sample(lang_df)
    train_df = pd.concat((train_df, train_lang))
    val_df = pd.concat((val_df, val_lang))
```

```python
train_df.reset_index(inplace=True)
```

```python
val_df.reset_index(inplace=True)
```
