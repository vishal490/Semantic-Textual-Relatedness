## Load Pre Trained Model

```python
pretrained_model = SentenceTransformer('sentence-transformers/LaBSE', device="cuda")
```

```python
count = 0
def compute_similarity_labse(row_data, model):

    global count

    count += 1
    print(count)
    text1, text2 = row_data["Text1"], row_data["Text2"]

    text1_embedding, text2_embedding = model.encode([text1, text2])

    # text1_embedding, text2_embedding = model_e5.encode([text1, text2])

    return 1-scipy.spatial.distance.cosine(text1_embedding, text2_embedding)

def compute_similarity_e5(row_data, model):

    global count

    count += 1
    print(count)
    text1, text2 = row_data["Text1"], row_data["Text2"]

    text1_embedding, text2_embedding = model.encode([
        f"query: {text1}", f"query: {text2}"])

    # text1_embedding, text2_embedding = model_e5.encode([text1, text2])

    return 1-scipy.spatial.distance.cosine(text1_embedding, text2_embedding)
```

## Evaluate over Training Data over Pretrained Model

```python
df_train = pd.read_csv("/content/drive/MyDrive/STR/train.csv")
```

```python
df_train["pretrained_scores"] = df_train.apply(lambda ele: compute_similarity_labse(ele, pretrained_model), axis=1)
```

```python
df_train.to_csv("/content/drive/MyDrive/STR/train_labse_responses.csv", index=False)
```
