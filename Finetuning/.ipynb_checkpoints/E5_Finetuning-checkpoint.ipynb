```python
from sentence_transformers import SentenceTransformer, LoggingHandler, losses, InputExample
from torch.utils.data import DataLoader
```

```python
model = SentenceTransformer('intfloat/multilingual-e5-large', device="cuda")
```

```python
# Create a cutoff based on 0.70 score
df_tr["Labels"] = df_tr.Scores.apply(lambda ele: 1 if ele > 0.5 else 0)
df_tr
```

```python
df_tr.Lang.unique()
```

```python
df_tr.Labels.value_counts()
```

```python
combined = pd.DataFrame()
for lang, group_df in df_tr.groupby(by="Lang"):
  if lang != "eng":
    combined = pd.concat((combined, group_df))
  else:
    combined = pd.concat((combined, group_df.sample(n=1700)))
```

```python
df_tr = combined

for lang, group_df in df_tr.groupby(by="Lang"):
  print(lang, group_df.shape)
```

```python
train_examples = []
for obj in df_tr[["Text1", "Text2", "Scores"]].to_dict(orient="records"):
    inp = InputExample(texts=[obj["Text1"], obj["Text2"]], label=obj["Scores"])
    train_examples.append(inp)
```

```python
len(train_examples)
```

```python
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=4, num_workers=2)
```

```python
train_loss = losses.ContrastiveLoss(model=model)
```

```python
model.fit([(train_dataloader, train_loss)], show_progress_bar=True, epochs=5, output_path="/content/drive/MyDrive/STR/e5_model_v8_5_epoch_04_11", steps_per_epoch=500)
```

```python
model_latest = SentenceTransformer("./drive/MyDrive/STR/e5_model_v8_5_epoch_04_11")
```

## Check the results now

```python
import scipy
count = 0
```

```python
def compute_similarity_e5(row_data, model):

    global count

    count += 1
    print(count)
    text1, text2 = row_data["Text1"], row_data["Text2"]

    text1_embedding, text2_embedding = model.encode([
        f"query: {text1}", f"query: {text2}"])

    # text1_embedding, text2_embedding = model_e5.encode([text1, text2])

    return 1-scipy.spatial.distance.cosine(text1_embedding, text2_embedding)
```

```python
df_val = pd.read_csv("/content/drive/MyDrive/STR/val.csv")
df_val
```

```python
df_val[df_val.Lang == "eng"]
```

```python
df_val.to_csv("/content/drive/MyDrive/STR/e5_val.csv", index=False)
```

```python
df_val["e5_trained_full"] = df_val.apply(lambda ele: compute_similarity_e5(ele, model_latest), axis=1)
```

```python
df_val["Labels"] = df_val.Scores.apply(lambda ele: 1 if ele > 0.5 else 0)
df_val["Labels_without_training"] = df_val.e5_mean.apply(lambda ele: 1 if ele > 0.5 else 0)
df_val["Labels_with_training_full"] = df_val.e5_trained_full.apply(lambda ele: 1 if ele > 0.5 else 0)
```
